# Introduction

After the meeting of 28/04/2021, we decided to split the PRIMAP-HIST dataset (with extrapolation) into 2 datasets depending on the 2 different scenarios.

The first one will contain all the data with the HISTCR scenario and the other, the HISTTP scenario.

# Dataset loading

```python
import pandas as pd

folder = "../../../data/ghg-emissions/primap-hist/"
dataset_name = "PRIMAP-hist_v2.2_19-Jan-2021.csv"
primap_hist_data = pd.read_csv(folder+dataset_name)
```

```python
primap_hist_data.info()
```

# Dataset splitting

Getting all the HISTCR scenario data.

```python
primap_histcr_data = primap_hist_data[primap_hist_data.scenario == "HISTCR"]
primap_histcr_data.shape
```

```python
primap_histcr_data.head()
```

```python
primap_histtp_data = primap_hist_data[primap_hist_data.scenario == "HISTTP"]
primap_histtp_data.shape
```

```python
primap_histtp_data.head()
```

# Datasets exports

```python
primap_histcr_data.to_csv(folder+'PRIMAP-HISTCR.csv')
```

```python
primap_histtp_data.to_csv(folder+'PRIMAP-HISTTP.csv')
```
